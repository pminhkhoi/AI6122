```python
import PySimpleGUI as sg
from whoosh import index
from whoosh import query
import whoosh
from whoosh import qparser

sg.theme('Reddit')
ConList=['SIGIR', 'WSDM', 'CIKM', 'SIGMOD', 'VLDB', 'ICML', 'AAAI', 'ACL', 'EMNLP']
# YearList=list(doc.keys())
YearList=list(range(2011, 2024))

layout = [  [sg.Text('Title:'), sg.InputText()],
            [sg.Text('Author:'), sg.InputText()],
            [sg.Text('Conference:'),sg.InputText(),sg.Text('Year:'),sg.InputText()],
            [sg.Button('Ok'), sg.Button('Cancel')] ]

ix = index.open_dir("indexdir", indexname="dblp")

QueryList=[]
outputstr=""
# Create the Window
window = sg.Window('Search App', layout)
# Event Loop to process "events" and get the "values" of the inputs
while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED or event == 'Cancel': # if user closes window or clicks cancel
        break
    print('You entered ', values[0],values[1],values[2],values[3])
    print("(",values[0],")")
    if values[0]=="":
        pass
    else:
        QueryList.append(qparser.QueryParser("title", schema=ix.schema).parse(values[0]))
    if values[1]=="":
        pass
    else:
        QueryList.append(qparser.QueryParser("author", schema=ix.schema).parse(values[1]))
    if values[2]=="":
        pass
    else:
        QueryList.append(qparser.QueryParser("publication_venue", schema=ix.schema).parse(values[2]))
    if values[3]=="":
        pass
    else:
        QueryList.append(qparser.QueryParser("year", schema=ix.schema).parse(values[3]))
    print(type(QueryList[0]))
    boolean_query = query.And(QueryList)
    print(boolean_query)
    with ix.searcher() as searcher:
    # Create a results object to hold the top N results
        results = searcher.search(boolean_query, limit=10)
        # Print the rank, score, docID, and snippet for each result
        if results.scored_length() > 0:
            for hit in results:
                print(hit)
                print(f"Rank: {hit.rank+1}, Score: {hit.score}, DocID: {hit.docnum}")
        else:
            print("There is no match for your query. Please try something else.")
    # results = searcher.search(boolean_query)
    # for result in results:
    #     print(result.docnum)
    #     outputstr+=str(result.docnum)
    sg.popup_scrolled(outputstr)
window.close()
```
